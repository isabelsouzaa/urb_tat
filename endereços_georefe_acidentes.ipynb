```python
from geopy.geocoders import Nominatim
```

```python
locator = Nominatim(user_agent="myGeocoder")
```

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
end_2016 =pd.read_excel('C:/Users/User/Desktop/DOUTORADO/URBANA/Paper/Urbano - REC/Dados/Acidentes/ACIDENTES PREF REC/enderec_p_python/enderc_2016.xlsx')
```

```python
end_2016.head()
```

```python
end_2016.dropna(inplace = True)
```

```python
def corrigir_num(numer):
    numer = numer.replace('.0', '')
    return numer
```

```python
before = end_2016.dtypes
```

```python
end_2016["numero"] = end_2016["numero"].astype(str)
```

```python
end_2016["bairro"] = end_2016["bairro"].astype(str)
```

```python
end_2016["endereco"] = end_2016["endereco"].astype(str)
```

```python
end_2016['numero'] = end_2016['numero'].apply(corrigir_num)
```

```python
end_2016.head()
```

```python
end_2016['georef'] = end_2016['bairro'] + ", " +end_2016['endereco'] + ", " + end_2016['numero']
```

```python
def corrigir_nomes(nome):
    nome = nome.replace('.', '').replace('ç', 'c').replace('ô', 'o').replace('é', 'e').replace('Á','A').replace('Â','A').replace('Ã','A')
    return nome
```

```python
end_2016['georef'] = end_2016['georef'].apply(corrigir_nomes)
```

```python
end_2016.head()
```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
end_2016['geo_loc']= end_2016['georef'].apply(geocode)
```

```python
end_2016['geo_loc']
```

```python
end_2016 ['latitude'] = end_2016['geo_loc'].apply(lambda loc: loc.latitude if loc else None)
end_2016 ['longitude'] = end_2016['geo_loc'].apply(lambda loc: loc.longitude if loc else None)
```
